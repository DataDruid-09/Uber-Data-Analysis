#Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("UberDataset.csv")
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

#Data Preprocessing


```python
df["PURPOSE"].fillna("Not", inplace = True)
```

```python
df.head()
```

```python
df["START_DATE"] = pd.to_datetime(df["START_DATE"], errors = "coerce")

df["END_DATE"] = pd.to_datetime(df["END_DATE"], errors = "coerce")
```

```python
df.info()
```

```python
from datetime import datetime

df["DATE"] = pd.DatetimeIndex(df["START_DATE"]).date
df["TIME"] = pd.DatetimeIndex(df["START_DATE"]).hour
```

```python
df.head()
```

```python
df.head()
```

```python
df["DAY-NIGHT"] = pd.cut(x=df["TIME"],bins = [0, 10, 15, 19, 24], labels = ["Morning", "Afternoon", "Evening", "Night"])
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

#Data Visulization

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)

sns.countplot(df["CATEGORY"])
plt.xticks(rotation = 90)

plt.subplot(1,2,2)
sns.countplot(df["PURPOSE"])
```

```python
sns.countplot(df["DAY-NIGHT"])
```

```python
df.head()
```

```python
df["MONTH"] = pd.DatetimeIndex(df["START_DATE"]).month

month_label = {1.0: "Jan", 2.0: "Feb", 3.0: "Mar", 4.0: "Apr", 5.0: "May",
               6.0: "Jun", 7.0: "Jul", 8.0: "Aug", 9.0: "Sep", 10.0: "Oct",
               11.0: "Nov", 12.0: "Dec"}

df["MONTH"] = df.MONTH.map(month_label)

mon = df.MONTH.value_counts(sort=False)
```

```python
df.head()
```

```python
dataframe = pd.DataFrame({
    "MONTHS": mon.values,
    "VALUE COUNT": df.groupby("MONTH", sort=False)["MILES"].max()
})

p = sns.lineplot(data=dataframe)
p.set(xlabel="MONTHS", ylabel="VALUE COUNT")
```

```python
df.head()
```

```python
df["DAY"] = df.START_DATE.dt.weekday

data_label = {
    0: 'Mon', 1: "Tue", 2: "Wed", 3: "Thu", 4: "Fri", 5: "Sat", 6: "Sun"
}
df["DAY"] = df["DAY"].map(data_label)
```

```python
df.head()
```

```python
data_label = df.DAY.value_counts()

sns.barplot(x=data_label.index, y=data_label)
plt.xlabel("DAY")
plt.ylabel("COUNT")
```
